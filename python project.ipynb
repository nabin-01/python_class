```python
from bs4 import BeautifulSoup
import requests
url = BeautifulSoup('https://www.worldometers.info/coronavirus/', 'html.parser')
response = requests.get(url)
response
```

```python
soup =response.text
soup = BeautifulSoup(soup,"lxml")
soup
```

```python
soup.table
```

```python
tags=soup.find_all("tr")
tags
```

```python
for i in tags:
    print(i.text)
```

```python
data=[]
for i in tags:
    data.append(i.text)
data

```

```python
datas=data[0:25]
print(datas)
```

```python
final_output=[]
for i in datas:
    a=i.split('\n')[0:11]
    final_output.append(a)
final_output
```

```python
import csv
with open('world_meter.csv', 'a') as f:
    x = csv.writer(f)
    for i in final_output:
        x.writerow(i)
f.close()
```
