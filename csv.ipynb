```python
#DictWriter in csv module
import csv
csv_columns = ['name','age','address']
dict = [{'name':'ABCD','age':18,"address":'BKT'},{'name':"Shyam",'age':40,'address':'KTM'}]
try:
    with open('data3.csv','w') as csvfile:
        writer = csv.DictWriter(csvfile,fieldnames = csv_columns)
        writer.writeheader()
        for data in dict:
            writer.writerow(data)
except IOError:
    print('IOError.')
```

```python
#DictWriter for csv module
import csv
csv_columns = ['Name','Age','Address']
dict = [{'Name':'Aayush','Age': 50,'Address':'KTM'},{'Name':'Shyam','Age':60,'Address': 'Patan'}]
try:
    with open('data.csv','w') as csvfile:
        writer = csv.DictWriter(csvfile,fieldnames = csv_columns)
        writer.writeheader()
        for data in dict:
            writer.writerow(data)
except IOError:
    print('IOError')
```
